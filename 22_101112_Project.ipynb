```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df = pd.read_csv("Studying with friends.csv")
df.head()
```

```python
df.shape
```

```python
df = df.drop("Timestamp",axis = 1)
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
# All null values where in the same row thus we will drop it
df.dropna(inplace=True)
df.isna().sum()
```

```python
df.columns
```

```python
plt.bar(df['Gender '].unique(),df['Gender '].value_counts(),color = ['deepskyblue','deeppink'])
plt.title('Gender Distribution', fontweight='bold')
plt.ylabel('Count')
plt.show()
```

```python
plt.bar(df['On average, how frequently do you study with friends?'].unique(),df['On average, how frequently do you study with friends?'].value_counts(),color = ['teal','darkturquoise','deepskyblue','lightskyblue'])
plt.title('Studying With Friends Frequency Distribution', fontweight='bold')
plt.ylabel('Count')
plt.show()
```

```python
plt.bar(df['In your opinion, does studying with friends enhance your understanding of the study material'].unique(),df['In your opinion, does studying with friends enhance your understanding of the study material'].value_counts(),color = ['darkgreen','green','mediumseagreen','aquamarine','turquoise'])
plt.title('Does Studying With Friends Enhances Understanding?', fontweight='bold')
plt.ylabel('Count')
plt.show()
```

```python
plt.bar(df['When studying with friends, do you feel it helps you stay focused?'].unique(),df['When studying with friends, do you feel it helps you stay focused?'].value_counts(),color = 'orange')
plt.title('Rating of overall Impact of Studying with Friends', fontweight='bold')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

```python
# mean
mean = np.mean(df["On average, How many hours do you spend studying with friends?"])
print(f'On average EUI students spend {mean} hours studying with friends.')
# median
median = np.median(df["On average, How many hours do you spend studying with friends?"])
print(f'The median of hours EUI students spend studying with friends is: {median}')
# mode
mode = stats.mode(df["On average, How many hours do you spend studying with friends?"])
print(f'{mode[1]} EUI students study for {mode[0]} with friends.')
```

```python
plt.hist(df["On average, How many hours do you spend studying with friends?"])
plt.title("Distribution of hours spent studying with friends")
plt.xlabel("Hours")
plt.ylabel("Count")
plt.show()
```

```python
df.boxplot(return_type='axes')
```

```python
array=np.zeros(74)
for i in range(len(df["On average, how frequently do you study with friends?"])):
    if df["On average, how frequently do you study with friends?"][i] == "Never":
        array[i] = 0
    elif df["On average, how frequently do you study with friends?"][i] == "Rarely":
        array[i] = 1
    elif df["On average, how frequently do you study with friends?"][i] == "Often":
        array[i] = 2
    elif df["On average, how frequently do you study with friends?"][i] == "Always":
        array[i] = 3
np.corrcoef(array,df["On average, How many hours do you spend studying with friends?"])[1][0]
```

```python
plt.scatter(df["On average, How many hours do you spend studying with friends?"],array,color='green',alpha = 0.7)
plt.xlabel("Hours spent studying with friends")
plt.ylabel("Frequency")
plt.title("Scatter Plot of Hours spent studying with friends and Frequency")
plt.grid(True, linestyle='-', alpha=0.5)
plt.show()
```

```python
array2 =np.zeros(74)
for i in range(len(df['In your opinion, does studying with friends enhance your understanding of the study material'])):
    if df['In your opinion, does studying with friends enhance your understanding of the study material'][i] == "Strongly disagree":
        array2[i] = 0
    elif df['In your opinion, does studying with friends enhance your understanding of the study material'][i] == "Disagree":
        array2[i] = 1
    elif df['In your opinion, does studying with friends enhance your understanding of the study material'][i] == "Neutral":
        array2[i] = 2
    elif df['In your opinion, does studying with friends enhance your understanding of the study material'][i] == "Agree":
        array2[i] = 3
    elif df['In your opinion, does studying with friends enhance your understanding of the study material'][i] == "Strongly agree.":
        array2[i] = 4
np.corrcoef(array2,df["On average, How many hours do you spend studying with friends?"])[1][0]
```

```python
plt.scatter(df["On average, How many hours do you spend studying with friends?"],array2,color='green',alpha = 0.7)
plt.xlabel("Hours spent studying with friends")
plt.ylabel("Effect of studying wiht freinds")
plt.title("Scatter Plot of Hours spent studying with friends and it's effect")
plt.grid(True, linestyle='-', alpha=0.5)
plt.show()
```
